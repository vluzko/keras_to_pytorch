```python
import pandas as pd
from pathlib import Path
import seaborn
%cd /home/vincent/projects/keras_to_pytorch/data/
```

```python
dfs = {str(x)[8: -4] : pd.read_csv(str(x), index_col=0) for x in Path(".").glob("*.csv")}
```

```python
df = dfs['HLA-A*01:01']
print(type(df.allele))
seaborn.lmplot('torchMHC_prediction', 'MHCFlurry_prediction', data=df, fit_reg=False, scatter=True)
```

```python
all_dfs = pd.concat(list(dfs.values()))
# print(list(dfs.values())[0])
```
